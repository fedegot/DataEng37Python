```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal")

plt.title("fMRI Signal Across Time")# Assign a title
plt.xlabel("Time Point") #update the axes
plt.ylabel("Percent Signal Change") #update the axes
```

```python
sns.lineplot(
    data=fmri,
    x="timepoint",
    y= "signal",
    hue="event",
    palette=["teal", "indigo"]
)
```

```python
sns.relplot(
    data=fmri, 
    x="timepoint",
    y="signal",
    hue="event",
    col="region",
    kind="line"
)
```

```python
#SEABORN
```

```python
data = fmri[["timepoint", "signal"]].groupby("timepoint").mean()
data.head()
```

```python
plt.plot(data)
plt.title("Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("% Signal Chance")
plt.show()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
plt.scatter(iris["petal_length"], iris["petal_width"], marker=".")
plt.xlabel("Petal Length")
plt.ylabel("Pedal Width")
plt.show()
```

```python
vers = iris[iris["species"] == "versicolor"]
virg = iris[iris["species"] == "virginica"]
seto = iris[iris["species"] == "setosa"]

plt.scatter(vers.petal_length, vers.petal_width, marker="o")
plt.scatter(virg.petal_length, virg.petal_width, marker="x")
plt.scatter(seto.petal_length, seto.petal_width, marker=".")

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend(labels=["Versicolor", "Virginica", "Setosa"])
plt.title("For Different Special of Iris")
```

```python
sns.pairplot(iris, hue="species")
```

```python
sns.relplot(
    data= iris,
    x="petal_length",
    y="petal_width",
    hue="species"
).set(
    title="Iris Petals",
    xlabel="Length",
    ylabel="Width"
)
```

```python
sns.lmplot(data=iris, x="petal_length", y="petal_width", hue="species")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
boarded = titanic.groupby("class").pclass.count()
boarded
```

```python
import squarify
```

```python
c = ["xkcd:blue","xkcd:pink","xkcd:teal",]
squarify.plot(boarded, label=boarded.index, color=c)
plt.axis("off")
```

```python
survived = titanic[titanic["alive"] == "yes"].groupby("class").pclass.count()
squarify.plot(survived, label=survived.index, color=c)
```
